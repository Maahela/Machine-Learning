```python
from sklearn.datasets import load_iris
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns


iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)

X.head()

# Try clustering into 3 groups
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Add cluster predictions to DataFrame
X['cluster'] = kmeans.labels_
X.head()

# Plot using 2 features for simplicity
sns.scatterplot(data=X, x='petal length (cm)', y='petal width (cm)', hue='cluster', palette='Set1')
plt.title("K-Means Clustering")
plt.show()

X['true_label'] = iris.target
pd.crosstab(X['cluster'], X['true_label'])
```
